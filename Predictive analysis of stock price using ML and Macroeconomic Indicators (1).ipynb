```python
# Feature Engineering 

df = pd.DataFrame(index = data.index)

# Lag features
df['MS(t-1)'] = data['MS'].shift(1)
df['JPM(t-1)'] = data['JPM'].shift(1)
df['BAC(t-1)'] = data['BAC'].shift(1)
df['C(t-1)'] = data['C'].shift(1)
df['WFC(t-1)'] = data['WFC'].shift(1)
df['SPY(t-1)'] = data['SPY'].shift(1)

# Lag features for Macreconomic Data
df['^VIX(t-1)'] = data['^VIX'].shift(1)
df['^TNX(t-1)'] = data['^TNX'].shift(1)
df['DX-Y.NYB(t-1)'] = data['DX-Y.NYB'].shift(1)
df['CL=F(t-1)'] = data['CL=F'].shift(1)
df['GC=F(t-1)'] = data['GC=F'].shift(1)

# Technical Inicators
df['C_MA5'] = data['C'].rolling(window=5).mean().shift(1)
df['C_MA10'] = data['C'].rolling(window=10).mean().shift(1)
df['C_Volatility5'] = data['C'].pct_change(5).shift(1)

df = df.dropna()

df['Target'] = data['C']
df
```

```python
df.isnull().sum()
```

```python
# Training the ML Algorithm

X = df.drop('Target', axis = 1)
Y = df['Target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, shuffle = False, test_size = 0.10)
```

```python
# Applying Decision Tree

from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(max_depth = 5)
dt_model.fit(X_train, Y_train)
```

```python
# Applying Random Forest
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators = 100)
rf_model.fit(X_train, Y_train)
```

```python
# Applying KNN 
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors = 5)
knn_model.fit(X_train, Y_train)
```

```python
# Applying SVM
from sklearn.svm import SVR
svr_model = SVR()
svr_model.fit(X_train, Y_train)
```

```python
dt_pred = dt_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
knn_pred = knn_model.predict(X_test)
svr_pred = svr_model.predict(X_test)
```

```python
result = pd.DataFrame(Y_test.index)
result['Actual'] = Y_test.values
result['DT Prediction'] = dt_pred
result['RF Prediction'] = rf_pred
result['KNN Prediction'] = knn_pred
result['SVR Prediction'] = svr_pred
result
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import pandas as pd

def evaluate_model(y_true, y_pred, model_name):
    r2 = r2_score(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)

    # Converting to pandas Series
    y_true = pd.Series(y_true)
    y_pred = pd.Series(y_pred)

    # Calculating direction changes
    actual_diff = y_true.diff().dropna().reset_index(drop=True)
    predicted_diff = y_pred.diff().dropna().reset_index(drop=True)

    # Making lengths match (if slightly off due to model trimming)
    min_len = min(len(actual_diff), len(predicted_diff))
    actual_diff = actual_diff[:min_len]
    predicted_diff = predicted_diff[:min_len]

    # MDA calculation
    mda = (np.sign(actual_diff) == np.sign(predicted_diff)).mean()

    print("Model name:", model_name)
    print("R2 Value:", r2)
    print("MSE:", mse)
    print("RMSE:", rmse)
    print("MDA:", round(mda, 4))
    print("\n")

evaluate_model(Y_test, dt_pred, "Decision Tree")
evaluate_model(Y_test, rf_pred, "Random Forest")
evaluate_model(Y_test, knn_pred, "K-Nearest Neighbor")
evaluate_model(Y_test, svr_pred, "Support Vector Regressor")
```

```python
# Random Forest
# Best overall performance: accurate in magnitude and decent in direction.

# Decision Tree
# Decent fit, but very poor at predicting direction.

# K-Nearest Neighbor
# Underperforms badly; likely not suitable for this task.

# Support Vector Regressor
# Very poor fit; only gets direction right, not the value.
```
